```python
'''
HOME ASSIGNMENT 1 - AUDIO ANALYSIS

Completed by Rashid Zaynutdinov
U2110075 | ICE 21-01
'''

import wave
import matplotlib.pyplot as plt
import numpy as np
from google.colab import drive

# Mount Google Drive (uncomment if the audio file is in your Drive)
# drive.mount('/content/drive', force_remount=True)

'''
PLEASE CONSIDER READING THE FOLLOWING COMMENT, IN CASE YOU WISH TO RUN THIS VERY PYTHON SCRIPT, LOCALLY ON YOUR PC

In order to run locally, remove the comment status from the two lines, listed below, the one that starts with SPECIFY.
Will allow you to run the given code on your desired directory
'''
# Specify the path to your audio file (replace with your actual path)

#data_folder = "/content/"  # Replace with your desired directory
#audio_file_path = data_folder + "original_audio.wav"  # Adjust the filename as needed


# Mount Google Drive (uncomment if the audio file is in your Drive)
# drive.mount('/content/drive', force_remount=True)

# Specify the path to your audio file (replace with your actual path)
audio_file_path = "/content/drive/MyDrive/Multimedia_Computing/original_audio.wav"        # Chosen path /content/drive/MyDrive/Multimedia Computing/original_audio.wav


# Check if using Google Drive (commented mounting section)
# If the path points to Google Drive, uncomment the mounting line
# and update `audio_file_path` accordingly (e.g., "/content/drive/MyDrive/audio.wav")

# Open the audio file
try:
  with wave.open(audio_file_path, "rb") as wav_file:
    # Get audio data
    frames = wav_file.readframes(wav_file.getnframes())
    # Convert frames to numpy array for easier manipulation
    audio_data = np.frombuffer(frames, dtype=np.int16)

    # Get sampling rate
    sample_rate = wav_file.getframerate()

    # Extract time axis (assuming single channel)
    time = np.arange(0, len(audio_data) / sample_rate, 1 / sample_rate)

    # Plot the waveform
    plt.figure(figsize=(10, 4))  # Set figure size
    plt.plot(time, audio_data)
    plt.xlabel("Time (seconds)")
    plt.ylabel("Amplitude")
    plt.title("Waveform of original_audio.wav")
    plt.grid(True)

    # Save the plot as PNG
    plt.savefig("1_audio_waveform.png")

    print("Waveform analysis completed and saved as 1_audio_waveform.png ")
except FileNotFoundError:
  print(f"Error: Audio file not found at {audio_file_path}")
```
